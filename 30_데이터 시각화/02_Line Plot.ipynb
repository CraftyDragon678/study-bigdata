# 2. Line Plot (선 그래프)
* 시간에 따른 변화의 추이를 시각화하는데 유용함
* 예) 주가 변동, 기온 변화, 아파트 가격 변동

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 2-1. 그래프 그리기

```python
# (1, 3), (8, 10) 두 점을 연결하는 선 그래프 그리기
x = [1, 8]
y = [3, 10]
plt.plot(x, y)
plt.show()
```

```python
# 두개의 점만 보여주기
plt.plot(x, y, 'o')
plt.show()
```

```python
# 점과 선을 둘다 그리기
plt.plot(x, y)
plt.plot(x, y, 'o')
plt.show()
```

```python
# (1, 2) ~ (10, 20)
x = np.arange(1, 11)
y = x * 2
plt.plot(x, y, marker='o')
plt.show()
```
