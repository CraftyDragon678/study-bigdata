```python
import pandas as pd
```

```python
queries = ['공연장', '노래방', '놀이동산', '도서관', '미술관', '영화관']
```

```python
def convert_tsv(location_type: str):
  df = pd.read_csv(f"./문화컨텐츠/{location_type}.csv", sep="\t")
  df['type'] = location_type
  return df
```

```python
df = pd.concat([convert_tsv(query) for query in queries])
df = df.reset_index(drop=True)
df.info()
df
```

```python
df['lng'] = df['lat']
df['lat'] = df['long']
final_df = df.drop('long', axis=1).iloc[:, [0, 1, 2, 3, 4, 6, 5, 7]]
```

```python
final_df.to_csv('./문화컨텐츠.csv', index=False)
```
