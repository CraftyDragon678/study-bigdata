```python
import pandas as pd
import os
import unicodedata
```

```python
kinds = ['연립', '오피스텔', '아파트', '단독']
files = {kind: [name for name in os.listdir("./부동산") if kind in unicodedata.normalize('NFC', name)] for kind in kinds}
```

```python
def get_one_df(file_name):
  df = pd.read_csv('./부동산/' + file_name, encoding='euc-kr', header=15, dtype={
    '번지': str, '본번': str, '부번': str,
  }, parse_dates=['해제사유발생일'])
  return df

def get_df(kind):
  df = pd.concat([
    get_one_df(file) for file in [
      name for name in os.listdir("./부동산")
        if kind in unicodedata.normalize('NFC', name)
    ]
  ]).reset_index(drop=True)
  df['종류'] = kind
  return df
```

```python
final_df = pd.concat([get_df(kind) for kind in kinds]).reset_index(drop=True)
```

```python
addresses = final_df[final_df['종류'] != '단독'].apply(lambda x: f"{x['시군구']} {x['번지']}", axis=1).drop_duplicates()
```

```python
address_fail_df = address_converted_df[address_converted_df['상태'].str.contains('실패')]
address_fail_df = address_fail_df.fillna('')
address_fail_df = address_fail_df.drop(['상태', '원본주소'], axis=1)
address_fail_df
```

```python
fails = [
  '고양덕양구', '고양일산동구', '고양일산서구',
  '성남분당구', '성남수정구', '성남중원구',
  '수원권선구', '수원영통구', '수원장안구', '수원팔달구',
  '안산단원구', '안산상록구',
  '안양동안구', '안양만안구',
  '용인기흥구', '용인수지구', '용인처인구',
  '창원마산합포구', '창원마산회원구', '창원성산구', '창원의창구', '창원진해구',
  '포항남구', '포항북구',
  '전주덕진구', '전주완산구',
  '천안동남구', '천안서북구',
  '청주상당구', '청주서원구', '청주청원구', '청주흥덕구',
]
def convert_fail(df, gu):
  df['원본주소'] = df['지번주소']
  df['지번주소'] = df['지번주소'].str.replace(gu, f"{gu[:2]}시 {gu[2:]}")
  return df
address_fail_resolve_df = pd.concat([convert_fail(address_fail_df[address_fail_df['지번주소'].str.contains(fail)], fail) for fail in fails]).reset_index(drop=True)
address_fail_resolve_df
```

```python
address_fail_resolve_df.to_excel('./부동산_해결.xlsx', index=False)
```
