```python
result = json.loads(json.dumps(temp))  # string -> dict
result
```

```python
# STATE_DT(기준일)
# DECIDE_CNT(확진자수)
# CLEAR_CNT(격리해제수)
# DEATH_CNT(사망자수)
# CARE_CNT(치료중환자수)
get_properties = lambda key: list(map(lambda x: x[key], result['response']['body']['items']['item']))
state_dts = get_properties('stateDt')
decide_cnts = get_properties('decideCnt')
clear_cnts = get_properties('clearCnt')
death_cnts = get_properties('deathCnt')
care_cnts = get_properties('careCnt')
```

```python
df_data = {
  '기준일': state_dts,
  '확진자수': decide_cnts,
  '격리해제수': clear_cnts,
  '사망자수': death_cnts,
  '치료중환자수': care_cnts,
}
df_data
```

```python
df = pd.DataFrame(data=df_data)
df
```
