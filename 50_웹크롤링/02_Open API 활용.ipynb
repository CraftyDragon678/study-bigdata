## 한국관광공사 - 관광정보서비스 (JSON)


```python
url = f'http://api.visitkorea.or.kr/openapi/service/rest/KorService/areaCode?ServiceKey={service_key}&numOfRows=10&pageNo=1&MobileOS=ETC&MobileApp=TestApp&_type=json'
url
```

```python
res = requests.get(url)
res
```

```python
data = res.json()
data
```

```python
num_of_rows = data['response']['body']['numOfRows']
total_count = data['response']['body']['totalCount']
num_of_rows, total_count
```

```python
query_count = math.ceil(total_count / num_of_rows)
query_count
```

```python
code = []
name = []
rnum = []

for page_no in range(1, query_count + 1):
  url = f'http://api.visitkorea.or.kr/openapi/service/rest/KorService/areaCode?ServiceKey={service_key}&numOfRows=10&pageNo={page_no}&MobileOS=ETC&MobileApp=TestApp&_type=json'
  res = requests.get(url)
  data = res.json()
  
  items = data['response']['body']['items']['item']

  for item in items:
    code.append(item['code'])
    name.append(item['name'])
    rnum.append(item['rnum'])

name
```

```python
df_data = {
  'name': name,
  'code': code,
  'rnum': rnum
}
pd.DataFrame(data=df_data)
```
