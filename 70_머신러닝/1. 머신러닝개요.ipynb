## 2-1. 머신러닝이란?

* ( 머신러닝 )은 규칙을 일일이 프로그래밍하지 않아도 자동으로 데이터에서 ( 규칙 )을 ( 학습 )하는 알고리즘을 연구하는 인공지능의 한 분야이다.

* 1959년 아서 사무엘은 기계 학습을 "기계가 일일이 코드로 명시하지 않은 동작을 데이터로부터 ( 학습 )하여 실행할 수 있도록 하는 알고리즘을 개발하는 연구 분야"라고 정의하였다.

## 2-2. 머신러닝의 분류

* 머신러닝에는 크게 ( 지도학습 ), ( 비지도학습 )이 있다.

* ( 지도학습 ) : 문제와 정답(레이블)을 모두 알려주고 학습시키는 방법

* ( 비지도학습 ) : 정답(레이블)이 없는 데이터를 비슷한 특징끼리 군집화하여 새로운 데이터에 대한 결과를 예측하는 방법

### 2-2-1. 지도학습 (Supervised Learning)

1) ( 분류 classification ) : 데이터가 입력되었을 때 몇 개의 클래스 중 하나로 분류하는 문제 ( 이진분류, 다중분류 )

2) ( 회귀 regression ) : 입력된 데이터에 대해 임의의 어떤 숫자를 예측하는 문제, 일반적으로 다루기 편하고 해석이 용이한 선형함수 형태를 취하며 선형회귀라고도 함 ( 단순선형회귀, 다중선형회귀 )

[참고] 독립변수 vs 종속변수

* ( 독립변수 ) : 다른 변수의 변화와 관계없이 독립적으로 변하고, 다른 변수의 값을 결정하는 변수

* ( 종속변수 ) : 독립변수의 변화에 따라 값이 변하는 변수

### 2-2-2. 비지도학습 (Un-Supervised Learning)

1) ( 군집화 clustering ) : 주어진 데이터 집합을 유사한 데이터들의 그룹으로 나누는  것

2) ( 차원축소 ) : 데이터가 가지고 있는 특성의 개수를 줄여나가면서 모델의 성능을 유지하는 방법

# 3. 머신러닝 구현하기

## 3-1. 프로젝트 수행 단계

1) 문제정의 : 지도학습 or 비지도학습, 분류 or 회귀  

2) 데이터 가져오기 : 데이터 훑어보기, 구조 파악하기

3) 데이터 탐색과 시각화 : 상관관계 조사

4) 학습을 위한 데이터 준비 : 데이터 정제하기(특성 제거, 채우기), 범주형 데이터 다루기, 특성 스케일링하기 (Min-Max 스케일링, 표준화)

5) 데이터 쪼개기 : 훈련 데이터와 테스트 데이터 세트를 분리하기

6) 모델 선택과 훈련 : 훈련세트에서 훈련하기

7) 교차 검증을 사용한 평가 : K-fold cross-validation

8) 모델 세부 튜닝 : 그리드 서치, 랜덤 서치

9) 테스트 세트로 시스템 평가하기

10) 런칭, 모니터링, 유지보수
